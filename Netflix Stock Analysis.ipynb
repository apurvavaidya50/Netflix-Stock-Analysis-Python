```python
import numpy as np  # Data Cleaning and Preprocessing
import pandas as pd
import matplotlib.pyplot as plt ## Visualisation
import seaborn as sns
from datetime import datetime  # To have date column in date format
```

```python
df = pd.read_csv("Netflix.csv")
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)}) #to set the default figure size for all subsequent Seaborn plots to be 10 inches wide and 5 inches tall
```

```python
df['Date'] =pd.to_datetime(df['Date'])  #To get Date in datetime format
df = df.set_index('Date') #To set Date column as index
df.head()
```

```python
 # 1. Volume of Stock Traded
```

```python
sns.lineplot(x = df.index, y = df['Volume'], label = 'Volume') 
plt.title('Volume of Stock Vs Time')
```

```python
# 2. Netflix Stock Price - High, Open, Close
```

```python
df.plot(y =['High','Low','Open'], title = "Netflix Stock Price")
```

```python
# 3. Netflix Stock Price - Day, Month, Year Wise
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(3, figsize = (15,10)) #To plot multiple axis and subplots in single graph
df.groupby(df.index.day).mean().plot(y = 'Volume',ax=ax1,xlabel = 'Day')
df.groupby(df.index.month).mean().plot(y ='Volume',ax =ax2,xlabel = 'Month')
df.groupby(df.index.year).mean().plot(y ='Volume',ax = ax3,xlabel = 'Year')
```

```python
# 4. Top-5 Dates with Highest Stock Price
```

```python
a = df.sort_values(by ='High',ascending = False).head(5)
a['High']
```

```python
# 5. Top 5 Dates with Lowest Stock Price
```

```python
b = df.sort_values(by ='Low',ascending = True).head(5)
b['Low']
```
